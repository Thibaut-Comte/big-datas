```python
import pandas as pd 
import numpy as np 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
hp = './datas/price_availability.csv'
hl = './datas/listings_final.csv'
prices = pd.read_csv(hp, sep=';')
listing = pd.read_csv(hl, sep=';')
```

```python
prices = prices.groupby(['listing_id']).mean()
prices
```

```python
listing = listing.groupby('listing_id').mean()
listing
```

```python
full = pd.merge(prices[['local_price']], listing, on=['listing_id'])
full
```

```python
# y prices
# x listing => person_capactity, bedrooms
y = np.array(full[['local_price']])
x = np.array(full[['person_capacity', 'bedrooms', 'bathrooms']])
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
y_test
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
reg = linear_model.LinearRegression().fit(x_train, y_train)
reg
```

```python
print("Coefficients beta_j :\n{}\nCoefficients INTERCEPT beta0 :\n{}".format(reg.coef_,reg.intercept_))
```

```python
y_pred = reg.predict(x_test)
y_pred
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
